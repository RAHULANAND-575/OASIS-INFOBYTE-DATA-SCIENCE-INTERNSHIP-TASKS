## Plots of the Regions 

```python
sns.countplot(x = df['Region'], palette = "dark")
plt.xticks(rotation ='vertical')
plt.show()
```

## Heatmap for Rate percentage

```python
plt.figure(figsize = (9,9))
sns.heatmap(df[['Estimated Unemployment Rate (%)',
                'Estimated Employed',
                'Estimated Labour Participation Rate (%)']].corr(),annot=True)
plt.show()
```

```python
# Box Plot of Estimated Unemployment Rate
plt.figure(figsize=(24, 14))
sns.boxplot(x = 'Region', y= 'Estimated Unemployment Rate (%)', data = df)
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
sns.pairplot(df)
```

```python
# Group by Region and calculate the mean of 'Estimated Unemployment Rate (%)'
x = df.groupby("Region")["Estimated Unemployment Rate (%)"].mean().sort_values(ascending=False).to_frame()

# Plotting
plt.figure(figsize=(15, 15))
sns.barplot(x=x.index, y=x["Estimated Unemployment Rate (%)"], palette=sns.color_palette("Greens", n_colors=len(x.index))[::-1])
plt.title("State-wise Average Unemployment Rate", fontweight="black", fontsize=20, pad=10)
plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate (%)')
plt.show()
```
